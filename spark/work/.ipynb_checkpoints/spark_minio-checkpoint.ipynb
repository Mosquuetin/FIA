```python
df = (
    spark.read.format("json").load("s3a://raw/Previsao/")
)

df.show()

df_exploded = df.select(
    "hr",
    "p.cp",
    "p.np",
    "p.px",
    "p.py",
    explode("p.l").alias("l_exploded")
    )

# Selecionar campos individuais da struct explodida
df_exploded = df_exploded.select(
    "hr",
    "cp",
    "np",
    "px",
    "py",
    "l_exploded.c",
    "l_exploded.cl",
    "l_exploded.sl",
    "l_exploded.lt0",
    "l_exploded.lt1",
    "l_exploded.qv",
    explode("l_exploded.vs").alias("l_exploded")
    )

df_exploded = df_exploded.select(
    "hr",
    "cp",
    "np",
    "px",
    "py",
    "c",
    "cl",
    "sl",
    "lt0",
    "lt1",
    "qv",
    "l_exploded.p",
    "l_exploded.t",
    "l_exploded.a",
    "l_exploded.ta",
    col("l_exploded.py").alias("py_vs"),
    col("l_exploded.px").alias("px_vs"),
    "l_exploded.sv",
    "l_exploded.is",
    )

df_exploded = df_exploded.dropDuplicates()

df_exploded.show()


df_exploded.write.format("parquet").mode("overwrite").save("s3a://refined/Previsao/")

print("Arquivo 'Previsao' salvo com sucesso")
```

```python
df = spark.read.format("parquet").load("s3a://refined/Previsao/")

lista_cl = df.select("cl").distinct().rdd.flatMap(lambda x: x).collect()

schema = StructType([
    StructField("linha", IntegerType(), True),
    StructField("veiculo", IntegerType(), True),
    StructField("media_diferenca", IntegerType(), True),
    StructField("media_minutos", StringType(), True)
])
    
    
df_final = spark.createDataFrame([], schema)

for cl in lista_cl:
    
    df_cl = df.filter(col("cl") == cl)
    
    lista = df_cl.select("p").distinct().rdd.flatMap(lambda x: x).collect()
    
    for item in lista:
        print(cl, item)
        
        df_filter = (
            df_cl.filter(col("p") == item)
            .dropDuplicates(["cp"])
            .select("hr","cp","p","np","lt0","lt1","t","a")
            .orderBy("t")
        )
        
        if df_filter.count() <= 1:
            pass
        else:
            window_spec = Window.orderBy("minutos")
            
            df_filter = (
                df_filter.withColumn("minutos", expr("CAST(SPLIT(t, ':')[0] AS INT) * 60 + CAST(SPLIT(t, ':')[1] AS INT)"))
                .withColumn("diff_minutes", col("minutos") - lag("minutos", 1).over(window_spec))
                .filter(col("diff_minutes").isNotNull())
            )
            
            df_media = (
                df_filter.agg({"diff_minutes": "avg"})
                .withColumnRenamed("avg(diff_minutes)", "media_diferenca")
                .withColumn("media_diferenca", round(col("media_diferenca"), 1))
                .withColumn("media_minutos",
                    concat(
                        lpad(floor(col("media_diferenca") % 60).cast("string"), 2, "0"),
                        lit(":"),
                        lpad(floor((col("media_diferenca") * 60) % 60).cast("string"), 2, "0")
            
                        )
                    )
                .withColumn("veiculo",lit(int(item)))
                .withColumn("linha",lit(cl))
                .select(
                    "linha",
                    "veiculo",
                    "media_diferenca",
                    "media_minutos"
                )
            )
            
            
            df_final = df_final.union(df_media)
        
df_final.show()

df_final.write.format("parquet").mode("overwrite").save("s3a://trusted/medias_linhas_veiculos/")

print("Arquivo 'medias_linhas_veiculos' salvo com sucesso")
```

```python
df = spark.read.format("parquet").load("s3a://refined/Previsao/")

lista_cl = df.select("cl").distinct().rdd.flatMap(lambda x: x).collect()

schema = StructType([
    StructField("linha", IntegerType(), True),
    StructField("hora", StringType(), True),
    StructField("media_onibus", IntegerType(), True)
])
    
    
df_final = spark.createDataFrame([], schema)

for cl in lista_cl:
    
    df_filter = (
        df.filter(col("cl") == cl)
        .dropDuplicates(["cp"])
        .select("cl","t","qv")
    )

    
    df_media = (
        df_filter.withColumn("hora", concat(substring(col("t"), 1, 2),lit(":00")))
        .groupBy("cl","hora").agg(avg("qv").alias("media_onibus"))
        .withColumn("media_onibus", round(col("media_onibus"), 0))
        .drop("t")
    ) 
            
    df_final = df_final.union(df_media).orderBy("linha","hora")

    
df_final.show()

df_final.write.format("parquet").mode("overwrite").save("s3a://trusted/medias_veiculos_horario/")

print("Arquivo 'medias_veiculos_horario' salvo com sucesso")
```
